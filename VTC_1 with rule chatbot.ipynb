```python
import re  #re là một thư viện trong Python được sử dụng để xử lý biểu thức chính quy.
import random
```

```python
class IOECHAT:
    ### potential negative  
    negative_responses = ("no", "nope", "nah", "naw", "không muốn", "sorry", "không", "không thể", "never")
    ### Exit conversation keywword
    exit_commands = ("bye", "quit", "goodbye", "later", "tạm biệt", "exit", "bb", "lần sau","hẹn gặp lại")
    # Random starter question -  Chatbot can start conversation with user
    random_questions = (
        "Chào bạn, mình là trợ lý ảo iTalk. Mình có thể giúp gì cho bạn không?\n",
        "Can I help you?\n",
        "Xin chào, mình là iTalk! Hãy trò chuyện nhé.\n",
        "Hello, I'm iTalk. Ask me any question.\n",
        "Chào bạn, rất vui khi được nói chuyện với bạn.\n",
        "Mình là trợ lý ảo iTalk của bạn! Bạn muốn mình giúp gì nào?\n",
        "Xin chào! Hãy trò chuyện nhé, đừng ngại!\n"
    )

    random_exit_command = ("Ok, have a nice day! \n",
                           "Okay, have a good day! \n",
                           "Ok, See you again! \n",
                           "It's nice talking to you. See you again soon. \n"
                          )
    
    def __init__(self):
        self.alienbabble = {'ok_answer': r'.*\s*ok.*',
                            'hello_answer': r'.*\s*hello.*',
                            'about_ioe_intent': r'.*\s*IOE là.*',
                            'about_the_le': r'.*\s*thể lệ.*',
                            'about_thanh_tich': r'.*\s*thành tích.*',
                            'about_doi_tuong': r'.*\s*đối tượng.*',
                            'about_sign_in': r'.*\s*đăng kí tài khoản.*',
                            'about_sign_in_student': r'.*\s*tài khoản học sinh.*'
                           }
        
    # def greet(self):
    #     self.greet_intent = input(random.choice(self.random_questions)).lower()
    #     if self.greet_intent in self.negative_responses or self.greet_intent in self.exit_commands:
    #         print(random.choice(self.random_exit_command))
    #             return 
    #         else: 
    #             self.chat()

    def make_exit(self, reply):
        for command in self.exit_commands:
            if reply == command:
                print(random.choice(self.random_exit_command))
                return True

    def chat(self): 
        self.reply = input(random.choice(self.random_questions)).lower()
        if self.reply in self.negative_responses or self.reply in self.exit_commands:
            print(random.choice(self.random_exit_command))
            return True
        while not self.make_exit(self.reply):
             self.reply= input(self.match_reply(self.reply)).lower()
            
    def match_reply(self, reply):
        for key, value in self.alienbabble.items():
            intent = key
            regex_pattern = value
            found_match = re.match(regex_pattern, reply)
            if found_match and intent == 'ok_answer':
                return self.ok_answer()
            elif found_match and intent == 'hello_answer':
                return self.hello_answer()
            elif found_match and intent == 'about_ioe_intent':
                return self.about_ioe_intent()
            elif found_match and intent == 'about_the_le':
                return self.about_the_le()
            elif found_match and intent == 'about_thanh_tich':
                return self.about_thanh_tich()
            elif found_match and intent == 'about_doi_tuong':
                return self.about_doi_tuong()
            elif found_match and intent == 'about_sign_in':
                return self.about_sign_in()
            elif found_match and intent == 'about_sign_in_student':
                return self.about_sign_in_student()
        
        if not found_match:
            return self.no_match_intent()
    
    def ok_answer(self):
        responses = ("Yeah, keep going on! \n",
                     "Hãy tiếp tục trò chuyện cùng mình nhé! \n",
                     "Trò chuyện thêm nữa nào! \n"
                    )
        return "iTalk: " + random.choice(responses)
        
    def hello_answer(self):
        responses = ("Yeah, keep going on! \n",
                     "Hãy tiếp tục trò chuyện cùng mình nhé! \n",
                     "Trò chuyện thêm nữa nào! \n"
                    )
        return "iTalk: " + random.choice(responses)
        
    def about_ioe_intent(self):
        responses = ("IOE, viết tắt của Internet Olympiads of English, là một cuộc thi tiếng Anh trên Internet được tổ chức bởi Bộ Giáo dục và Đào tạo dành cho học sinh từ lớp 3 đến lớp 12 trên toàn quốc. Mục tiêu chính của cuộc thi là khuyến khích sự học tập và rèn luyện tiếng Anh cho học sinh. \n",
                     "IOE - Hấp dẫn - Mê say - Thấy ngay kết quả là cuộc thi Tiếng Anh trên Internet được triển khai từ năm học 2010-2011 theo sự chỉ đạo của Bộ Giáo dục và Đào tạo (Bộ GD&ĐT) và Công ty Cổ phần VTC Truyền thông Trực tuyến (VTC Online) thuộc Tổng Công ty VTC truyền thông đa phương tiện là đơn vị trực tiếp triển khai cuộc thi qua việc xây dựng nền tảng kỹ thuật, xây dựng nội dung. Các đơn vị giáo dục và học sinh có thể lựa chọn IOE như một sân chơi tự nguyện và một kênh đánh giá năng lực tiếng Anh của học sinh. \n",
                     "Olympic tiếng Anh trên Internet - Internet Olympiads of English (IOE) là sân chơi giúp khơi dậy niềm yêu thích tiếng Anh của các em học sinh trên khắp Việt Nam, từ đó có thể ứng dụng và sử dụng tiếng Anh vào các lĩnh vực khác nhau trong cuộc sống. \n Slogan của cuộc thi là: \t\t\tHấp dẫn - Mê say - Thấy ngay kết quả\n"
                    )
        return "iTalk: " + random.choice(responses)
        
    def about_the_le(self):
        link_1 = "www.ioe.vn"
        link_2 = "https://ioe.vn/the-le"
        responses = ("Mỗi một năm học có 4 vòng thi Olympic chính thức dành cho học sinh cấp tiểu học (TH) và học sinh trung học cơ sở (THCS); 3 vòng thi Olympic chính thức cho học sinh cấp phổ thông trung học (PTTH) và các vòng tự luyện cho học sinh. Các vòng tự luyện được Ban tổ chức cuộc thi cấp toàn quốc cập nhật hàng tuần trên trang web chính thức www.ioe.vn \n",
                     "Thể lệ thi chi tiết có thể xem tại: https://ioe.vn/the-le  \n"
                    )
        return "iTalk: " + random.choice(responses)
        
    def about_thanh_tich(self):
        responses = ("IOE đã nhận được sự hưởng ứng tham gia của 30,2 triệu học sinh đến từ hơn 26.000 nhà trường, phủ rộng khắp 705 quận huyện của 63 tỉnh thành trên toàn quốc. \n",
                     "Với nỗ lực không ngừng, IOE tự hào với những thành tựu và kết quả tích cực đóng góp cho sự phát triển của giáo dục Việt Nam: \n\t- Nhận giải thưởng Chuyển đổi số Việt Nam 2022 - Hạng mục “Sản phẩm, giải pháp chuyển đổi số vì cộng đồng” \n\t- Nhận giải thưởng Công nghệ Giáo dục Tiêu biểu 2022 tại Diễn đàn Giáo dục và Triển lãm Học đường 4.0 năm 2022 \n\t- Tiên phong sử dụng công nghệ AI hoàn toàn miễn phí giúp học sinh rèn luyện phát âm nâng cao kỹ năng nói tiếng Anh theo chuẩn quốc tế \n\t- Góp phần tích cực vào việc thực hiện “Đề án dạy và học ngoại ngữ trong hệ thống giáo dục quốc dân giai đoạn 2017-2025” \n"
                    )
        return "iTalk: " + random.choice(responses)
        
    def about_doi_tuong(self):
        responses = ("Học sinh phổ thông từ lớp 3 đến lớp 12 trên toàn quốc, nếu có điều kiện truy cập Internet và có nguyện vọng tham gia Olympic Tiếng Anh trực tuyến, đều có thể đăng ký dự thi IOE. \n",
                     "Học sinh trên toàn nước Việt Nam từ lớp 3 đến lớp 12 đều có thể tham gia cuộc thi IOE. \n"
                    )
        return "iTalk: " + random.choice(responses)
        
    def about_sign_in(self):
        
        return "iTalk: " + "bạn lựa chọn tài khoản muốn đăng kí:     \n"
        
    def about_sign_in_student(self):
        responses = "Đăng kí tài khoản học sinh gồm các bước sau:\n\tBước 1: Truy cập trang chủ IOE.vn, chọn chức năng đăng ký.\n\tBước 2: Điền đầy đủ thông tin trên trang đăng ký rồi click vào Nhấn để đăng ký \n\tBước 3: Bạn điền các thông tin cá nhân vào các ô trống và click vào “Cập nhật thông tin” \n"
        return "iTalk: " + responses

    def no_match_intent(self):
        responses = ("Please tell me more! \n",
                    "Can you tell me more! \n",
                    "I dont understand what you talk about, try again! \n",
                    "I see. Can you elaborate? \n")
        return "iTalk: " + random.choice(responses)


IOE_Chatbot = IOECHAT()
IOE_Chatbot.chat()

```
